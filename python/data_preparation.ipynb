```python
mig = pd.read_csv('UNmigrants_2017_4.csv', index_col=0, sep=",",thousands=',')
mig.head(30)

test = mig[['destination','Total']]
test.head(300)
x=test.dropna()
x.head(300)
mig.head()
```

```python
#melt the matrix
mig2 = pd.melt(mig, id_vars=["hpyth", "destination", "sortcode", "Total","Other.North", "Other.South"], var_name="country2", value_name="inflow_1to2")
#mig2.tail(30)
#mig2.count()
```

```python
#remove all nans in count
mig3 = mig2.dropna(subset=['inflow_1to2'])
mig3.count()
mig3.head()
```

```python
mig3.country2.value_counts()
mig3['country2']=mig3['country2'].replace("\.",' ', regex=True)
```

## merge data

```python
mig4=pd.merge(mig3, country, on='destination')
mig4.tail(200)
#mig4.count()
```

```python
mig5=pd.merge(mig4, country, left_on='country2', right_on='destination', how='inner')
mig5 = mig5.rename(columns={'destination_x':'country1'})
mig5.count()
mig5.head()
```
